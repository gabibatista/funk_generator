### Reading data

```python
import re, pandas as pd
```

```python
df = pd.read_csv('dataset.csv', quotechar='+', header=0)
df['lyric_txt'] = df['lyric'].apply(lambda d: re.sub('|', ', ', d))
df['lyric'] = df['lyric'].apply(lambda d: d.split('|'))
```

```python
df.count()
```

```python
df
```

### Extracting features from lyric
- Stemming

```python
import nltk
```

```python
stemmer = nltk.stem.RSLPStemmer()
```

```python
df['words'] = df['lyric_txt'].apply(lambda p: re.findall('(\w+)', p))
```

```python
df
```

```python
df['stemmed_words'] = df['words'].apply(lambda l: list(stemmer.stem(w) for w in l))
```

```python
df
```

- CountVectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer()
vectorizer
```

```python
df['cnt_vect'] = vectorizer.fit_transform(df['lyric_txt'])
```

- POS tagging

```python
import nlpnet
# nltk.download()
nlpnet.set_data_dir('/home/gabriela.batista/nltk_data/pos-pt')
```

```python
import numpy as np
# save np.load
np_load_old = np.load

# modify the default parameters of np.load
np.load = lambda *a,**k: np_load_old(*a, allow_pickle=True, **k)


### https://stackoverflow.com/questions/55890813/how-to-fix-object-arrays-cannot-be-loaded-when-allow-pickle-false-for-imdb-loa/56062555
```

```python
tagger = nlpnet.POSTagger()
```

```python
tagged_lyrics = []
for lyric in df['lyric']:
    tagged_lyric = []
    for line in lyric:
        tagged_lyric.append(tagger.tag(line))
    tagged_lyrics.append(tagged_lyric)
df['tagged_lyric'] = tagged_lyrics
```
